```python
import numpy as np
import os
```

```python

def load_eeg_data(folder_path):
    file_list = sorted(os.listdir(folder_path))  # Sort to maintain order
    data = []

    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)
        
        # Load the file, assuming one integer per line
        eeg_series = np.loadtxt(file_path, dtype=int) 
         
        
        data.append(eeg_series)

    return np.array(data)  # Shape: (100, 4097)
```

```python
# Example usage:
folder_path = "N"  # Replace with actual folder path
eeg_array = load_eeg_data(folder_path)
np.save("N.npy", eeg_array)
```

```python
def FRD(timeseries):
    """Build rank-frequency distribution."""

    values, counts = np.unique(timeseries, return_counts=True)
    freqs = np.sort(counts)[::-1]
    ranks = np.arange(1, len(freqs) + 1)
    return ranks, freqs
```

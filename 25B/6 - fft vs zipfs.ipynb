```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
dataF = np.load('dataF.npy')
dataN = np.load('dataN.npy')
dataO = np.load('dataO.npy')
dataS = np.load('dataS.npy')
dataZ = np.load('dataZ.npy')

folders = [dataZ, dataO, dataN, dataF, dataS]
filenames = ['Z', 'O', 'N', 'F', 'S']
df_filenames = ['Z.csv', 'O.csv', 'N.csv', 'F.csv', 'S.csv']

dfZ = pd.read_csv('Z.csv')
dfO = pd.read_csv('O.csv')
dfN = pd.read_csv('N.csv')
dfF = pd.read_csv('F.csv')
dfS = pd.read_csv('S.csv')

df_folders = [dfZ, dfO, dfN, dfF, dfS]
```

```python
dfN.shape
print(dfZ.shape[0])

time = np.linspace(0,23.6,4097)

dfZ['time'] = time
dfO['time'] = time
dfN['time'] = time
dfF['time'] = time
dfS['time'] = time

dfZ.head()
```

```python
f = 0
for i in df_folders:
    for n in range(100):
        fig, axs = plt.subplots(3, figsize=(40,30))
        axs[0].set_title(f'Time Series # {n}', fontsize=40)
        axs[0].plot(dfZ['time'],i[f'{n}'])
        axs[0].set_xlabel('Time (s)', fontsize=20)
        axs[0].set_ylabel(f'Amplitude ($\mu$V)', fontsize=20)

        # FFT
        fft_values = fft(i[f'{n}'].values)  # Compute FFT
        fft_magnitude = np.abs(fft_values)  # Magnitude (amplitude)

        sample_spacing = i['time'].iloc[1] - i['time'].iloc[0]  # Time step
        frequencies = fftfreq(4097, sample_spacing)
        positive_freqs = frequencies[:4097//2]
        positive_magnitude = fft_magnitude[:4097//2] * 2 / 4097  

        axs[1].set_title(f'FFT # {n}', fontsize=40)
        axs[1].plot(positive_freqs, positive_magnitude, 'r-', linewidth=1)
        axs[1].set_xlabel('Frequency (Hz)', fontsize=20)
        axs[1].set_ylabel(f'Amplitude ($\mu$V)', fontsize=20)


        # HISTOGRAM
        values = i[f'{n}'].values
        value_counts = pd.Series(values).value_counts()
        ranks = range(1, len(value_counts) + 1)
        frequencies = value_counts.values

        axs[2].set_title(f'MEDIAN CENTERED DISTRIBUTION # {n}', fontsize=40)
        axs[2].scatter(ranks, frequencies, color='black', s=2)
        axs[2].set_xlabel('Rank (s)', fontsize=20)
        axs[2].set_ylabel(f'Frequency', fontsize=20)
        # axs[2].set_xscale('log')
        axs[2].set_yscale('log')

        plt.savefig(f'data_vis_{filenames[f]}{1000+n}', dpi=200)
        plt.close(fig)
        del fft_values, fft_magnitude, value_counts, values, frequencies
    
    f = f + 1
```
